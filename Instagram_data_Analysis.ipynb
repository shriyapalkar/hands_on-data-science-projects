```python


# Calculate the total impressions from each source
home = data["From Home"].sum()
hashtag = data["From Hashtags"].sum()
explore = data["From Explore"].sum()
other = data["From Other"].sum()

# Create a list of labels and values for the pie chart
labels = ['Home', 'Hashtags', 'Explore', 'Other']
values = [home, hashtag, explore, other]
figure=px.pie(data,values=values,names=labels,title="Percentage of Impressions from Various Sources on Instagram",hole=0.5)
figure.show()
# Create the pie chart using matplotlib
plt.figure(figsize=(5, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Impressions from Various Sources on Instagram')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

The above plot shows that almost
44% of the reach from the followers,
34% of the reach from the hashtag,
19.2% of the reach from the explore and
13% of the reach from the others.

```python
#analyzing conversion rate
#In Instagram, conversation rate means how many followers you are getting from the number of profile visits from a post.
#The formula that you can use to calculate conversion rate is (Follows/Profile Visits) * 100.
conversion_rate = (data["Follows"].sum() / data["Profile Visits"].sum()) * 100
print(conversion_rate)
```

```python
#scatter plot
x=data['Profile Visits'] # Replace 'profilevisit' with 'Profile Visits'
y=data['Follows']
plt.scatter(x,y)
plt.xlabel("Profile Visits")
plt.ylabel("Follows")
plt.title("Relationship between Profile Visits and Follows")
plt.show()
```

```python
#prediction
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor
#Instagram reach prediction
x = np.array(data[['Likes', 'Saves', 'Comments', 'Shares',
                   'Profile Visits', 'Follows']])
y = np.array(data["Impressions"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y,
                                                test_size=0.2,
                                                random_state=64)
model = PassiveAggressiveRegressor()
model.fit(xtrain, ytrain)
model.score(xtest, ytest)
```

```python
#feature extraction

#let’s predict the reach of an Instagram post by giving inputs to the machine learning model:
# Features = [['Likes','Saves', 'Comments', 'Shares', 'Profile Visits', 'Follows']]
features = np.array([[282.0, 233.0, 4.0, 9.0, 165.0, 54.0]])
model.predict(features)
```

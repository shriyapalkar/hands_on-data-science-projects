By finding mode we have got 294 as the mode value.We will need to use this value for dividing the sum of all saples related to HDI,GDP and Population.

Lets create a new data set by combining the necessary columns from both dataset.

Aggregating the data

```python
code = df1["CODE"].unique().tolist()
country = df1["COUNTRY"].unique().tolist()
hdi = []
tc = []
td = []
sti = []
population = df1["POP"].unique().tolist()
gdp = []

for i in country:
  hdi.append((df1.loc[df1["COUNTRY"]==i,"HDI"]).sum()/294)
  tc.append((df.loc[df["location"]==i,"total_cases"]).sum())
  td.append((df.loc[df["location"]==i,"total_deaths"]).sum())
  sti.append((df1.loc[df1["COUNTRY"]==i,"STI"]).sum()/294)
  population.append((df.loc[df["location"]==i,"population"]).sum()/294)

aggregated_data=pd.DataFrame(list(zip(code,country,hdi,tc,td,sti,population)),
                               columns=["Country Code","Country","HDI","Total Cases","Total Deaths","Stringency Index","Population"])
print(aggregated_data.head())
```

We have not includes GDP_per_capita column because both the datasets doesn't have correct figure so we need manual data of the each country taken in dataset.

As it is very difficult to find gdp of all countries. we will select the sub sample from the data set.

To create sub sample we will be selecting top ten countries with highest number of covid 19 cases.So we have to sort accordingly.

```python
# Sorting Data According to Total Cases
```

```python
df1 = aggregated_data.sort_values(by=["Total Cases"], ascending=False)
print(df1.head())
```

```python
#total number of cases and total deaths in all these countries

fig = go.Figure()
fig.add_trace(go.Bar(x=df1["Country"],y=df1["Total Cases"],name='Total Cases',marker_color='indianred'))
fig.add_trace(go.Bar(x=df1["Country"],y=df1["Total Deaths"],name='Total Deaths',marker_color='lightsalmon'))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

```python
# Percentage of Total Cases and Deaths
cases = df1["Total Cases"].sum()
deceased = df1["Total Deaths"].sum()

labels = ["Total Cases", "Total Deaths"]
values = [cases, deceased]

fig = px.pie(df1, values=values, names=labels,
             title='Percentage of Total Cases and Deaths', hole=0.5)
fig.show()
```

```python
#death rate of covid 19
death_rate = (df1["Total Deaths"].sum() / df1["Total Cases"].sum()) * 100
print("Death Rate = ", death_rate)
```

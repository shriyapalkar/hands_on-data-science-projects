```python
df=pd.read_csv("/content/tips.csv")
df
```

```python
figure = px.scatter(data_frame = df, x="total_bill",
                    y="tip", size="size", color= "day", trendline="ols",
                    title='Tip vs total_bill')
figure.show()
```

```python
figure = px.scatter(data_frame = df, x="total_bill",
                    y="size", size="size", color= "day", trendline="ols",
                    title='total_bill vs gender')
figure.show()
```

```python
figure = px.scatter(data_frame = df, x="total_bill",
                    y="tip", size="size", color= "time", trendline="ols")
figure.show()
```

```python
figure = px.pie(df,
             values='tip',
             names='day',hole = 0.5)
figure.show()
```

```python
fig=px.bar(df,x='day',y='tip',
           color='tip',height=400)
fig.show()
```

```python
import plotly.express as px


fig = px.bar(df, x='day', y='tip', color='time', barmode='group',
             title='Tips by Day and Time',
             labels={'tip': 'Total Tip', 'day': 'Day of Week', 'time': 'Meal Time'})
fig.show()
```

```python
day_time_count = df.groupby(['day', 'time'])['tip'].sum().reset_index()

fig = px.bar(day_time_count,
    x='day',
    y='tip',
    color='time',
    barmode='group',
    title='Tips by Day and Time'
)
fig.show()
```

```python
day_time_count = df.groupby(['day', 'smoker'])['tip'].sum().reset_index()

fig = px.bar(day_time_count,
    x='day',
    y='smoker',
    color='day',
    barmode='group',
    title='Tips by Day and Time'
)
fig.show()
```

```python
fig = px.bar(df, x="day",y="smoker", color="sex", barmode="group",
                  title="Number of Male and Female Smokers")
fig.show()
```

```python
fig = px.bar(df, x='day', y='tip', color='time',
             barmode='group', title='Tips by Day and Time',
             color_discrete_map={'Lunch':'pink', 'Dinner':'yellow'})
fig.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['sex_binary'] = le.fit_transform(df['sex'])
print(df[['sex', 'sex_binary']].head())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['smoker_binary'] = le.fit_transform(df['smoker'])
print(df[['smoker', 'smoker_binary']].head())
```

```python
df['smoker']=df['smoker'].map({'No':0,'Yes':1})
df['smoker']
```

```python
df['sex']=df['sex'].map({'Female':0,'Male':1})
df['sex']
```

```python
df['time']=df['time'].map({'Lunch':0,'Dinner':1})
df['time']
df['day']=df['day'].map({'Thur':0,'Fri':1,'Sat':2,'Sun':3})
df['day']
```

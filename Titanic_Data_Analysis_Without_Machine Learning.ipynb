```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Titanic-Dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df
```

```python
#descriptive statistics
df.describe()
```

```python
df.describe()
#Frequency count of categorical variables
df['Embarked'].value_counts()
```

```python
df['Sex'].value_counts()
```

```python
df['Pclass'].value_counts()
```

```python
#EDA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#survival count
sns.countplot(data=df,x='Survived')
plt.title('Survival Count')
plt.show()
```

```python
#age distribution
sns.histplot(data=df,x='Age',bins=20)
plt.title('Age Distribution')
plt.show()
```

```python
# Age distribution
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
# kde = pdf
# kde represents data using continous probability density/distribution function/curve in onr or more dimension.
# kde is generally used for visualisation of observation in a dataset.
```

```python
# fair distribution using kde
sns.histplot(df['Fare'], kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# survival rate for p class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Pclass')
plt.show()
```

Key observations to look for:

Survival rate by gender: Often, women tend to have higher survival rates than men.
Survival rate by class (Pclass): Passengers in 1st class typically have a higher survival rate than those in 3rd class.
Age distribution: There may be a correlation between age and survival, such as children having higher survival rates.

```python
#correlation matrix
numerical_data = df.select_dtypes(include=['number'])

corr_matrix = numerical_data.corr()
sns.heatmap(corr_matrix, annot=True,cmap='coolwarm')
#coolwarm seaborn feature for colour combination
plt.title('Correlation Matrix')
plt.show()
```

```python
# Survival Rate Analysis by Multiple Features
```

```python
survival_by_class_sex = df.pivot_table('Survived',['Pclass', 'Sex'],aggfunc='mean')
print(survival_by_class_sex)
```

```python
# surviral rate by age group
df['Age_Group']= pd.cut(df['Age'], bins=[0,12,18,30,50,80],labels=['Children','Teens','Young Adults','Adult','Senior'])
survival_by_class_sex = df.pivot_table('Survived','Age_Group',aggfunc='mean')
print(survival_by_class_sex)
```

```python
# Stacked bar plot for survival by class and gender
survival_by_class_sex = df.pivot_table('Survived',['Pclass', 'Sex'],aggfunc='mean')
survival_by_class_sex.plot(kind='bar', stacked=True)
```

```python
pd.crosstab([df['Pclass'],df['Sex']],df['Survived'],margins=True).plot(kind='bar',stacked=True)
plt.title('Survived by Class and Gender')
plt.ylabel('Count')
plt.show()
#
```

```python
# Pairplot to explore relations between multiple numerical features
sns.pairplot(df[['Age','Fare']])
plt.show()
```
